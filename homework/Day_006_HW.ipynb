# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
ecdf_data = app_train['AMT_INCOME_TOTAL']
def cdf(ecdf_data):
  x = np.sort(ecdf_data)
  y = np.arange(1,len(x)+1)/len(x)
  return(x,y)

def plot_ecdf(ecdf_data,xlabel='Value',ylabel='ECDF',label='AMT_INCOME_TOTAL'):
  x,y = cdf(ecdf_data)
  
  plt.plot(x,y,label=label)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.legend(markerscale=4)
  plt.figure(figsize=(10,10))
 
#plt.xlabel('Value')
#plt.ylabel('ECDF')
#plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
#plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plot_ecdf(ecdf_data)
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
#plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
#plt.xlabel('Value (log-scale)')
#plt.ylabel('ECDF')

#plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

#plt.show()
print(app_train['REGION_POPULATION_RELATIVE'].describe())
ecdf_data = app_train['REGION_POPULATION_RELATIVE']
def cdf(ecdf_data):
  x = np.sort(ecdf_data)
  y = np.arange(1,len(x)+1)/len(x)
  return(x,y)

def plot_ecdf(ecdf_data,xlabel='Value',ylabel='ECDF',label='REGION_POPULATION_RELATIVE'):
  x,y = cdf(ecdf_data)
  
  plt.plot(x,y,label=label)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.legend(markerscale=4)
  plt.figure(figsize=(10,10))
  
plot_ecdf(ecdf_data)
plt.show()
# 最大值為少數極端值

print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())
ecdf_data = app_train['OBS_60_CNT_SOCIAL_CIRCLE']
def cdf(ecdf_data):
  x = np.sort(ecdf_data)
  y = np.arange(1,len(x)+1)/len(x)
  return(x,y)

def plot_ecdf(ecdf_data,xlabel='Value',ylabel='ECDF',label='OBS_60_CNT_SOCIAL_CIRCLE'):
  x,y = cdf(ecdf_data)
  
  plt.plot(x,y,label=label)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.legend(markerscale=4)
  plt.figure(figsize=(10,10))
  
plot_ecdf(ecdf_data)
plt.show()
# 最大值為少數極端值
```
