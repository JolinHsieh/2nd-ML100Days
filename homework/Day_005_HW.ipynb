```python
##平均、直方圖
amt_income_mean = app_train['AMT_INCOME_TOTAL'].mean()
amt_income_std = app_train['AMT_INCOME_TOTAL'].std()
amt_credit_mean = app_train['AMT_CREDIT'].mean()
amt_credit_std = app_train['AMT_CREDIT'].std()
print(amt_income_mean,amt_income_std,amt_credit_mean,amt_credit_std)
```

```python
country_name = app_train.NAME_CONTRACT_TYPE.unique()
#print(country_name)
c_mean_all = []

for country in country_name:
  item = app_train['NAME_CONTRACT_TYPE'] == country
  #print(app_train[item])
  c_mean = app_train[item]['AMT_CREDIT'].mean()
  c_mean_all.append(c_mean)
 
df = pd.DataFrame(c_mean_all,index=country_name)
df.plot(kind='bar')
```
